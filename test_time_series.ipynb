```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing

# Generate some example time series data
np.random.seed(0)
data = np.random.randn(100).cumsum() + 50

# Fit Simple Exponential Smoothing (SES)
ses_model = SimpleExpSmoothing(data).fit()
ses_forecast = ses_model.forecast(10)

# Fit Holt's Linear Trend Method
holt_model = ExponentialSmoothing(data, trend='add').fit()
holt_forecast = holt_model.forecast(10)

# Fit Damped Exponential Smoothing
damped_model = ExponentialSmoothing(data, trend='add', damped_trend=True).fit()
damped_forecast = damped_model.forecast(10)

# Combine the forecasts using simple arithmetic average
combined_forecast = (ses_forecast + holt_forecast + damped_forecast) / 3

# Plot the results
plt.plot(data, label='Observed')
plt.plot(np.arange(len(data), len(data) + 10), ses_forecast, label='SES Forecast')
plt.plot(np.arange(len(data), len(data) + 10), holt_forecast, label='Holt Forecast')
plt.plot(np.arange(len(data), len(data) + 10), damped_forecast, label='Damped Forecast')
plt.plot(np.arange(len(data), len(data) + 10), combined_forecast, label='Combined Forecast', linestyle='--')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Generate some example time series data
np.random.seed(0)
data = np.random.randn(100).cumsum() + 50

# Fit Simple Exponential Smoothing (SES) model
ses_model = SimpleExpSmoothing(data).fit(smoothing_level=0.2, optimized=False)
ses_forecast = ses_model.forecast(10)

# Plot the results
plt.plot(data, label='Observed')
plt.plot(np.arange(len(data), len(data) + 10), ses_forecast, label='SES Forecast', linestyle='--')
plt.legend()
plt.show()
```
